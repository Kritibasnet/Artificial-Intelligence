**Task 1 — Unit Conversion Program**

```python
def unit_converter():
    """
    Converts units for length, weight, and volume.

    Parameters:
    None (takes user input)

    Returns:
    float : converted value
    """

    try:
        print("Choose conversion type:")
        print("1. Length")
        print("2. Weight")
        print("3. Volume")

        choice = input("Enter choice (1/2/3): ")

        value = float(input("Enter the value to convert: "))

        if choice == "1":
            print("1. meters to feet")
            print("2. feet to meters")

            sub = input("Choose option: ")

            if sub == "1":
                result = value * 3.28084
                print("Result:", result, "ft")

            elif sub == "2":
                result = value / 3.28084
                print("Result:", result, "m")

        elif choice == "2":
            print("1. kg to pounds")
            print("2. pounds to kg")

            sub = input("Choose option: ")

            if sub == "1":
                result = value * 2.20462
                print("Result:", result, "lbs")

            elif sub == "2":
                result = value / 2.20462
                print("Result:", result, "kg")

        elif choice == "3":
            print("1. liters to gallons")
            print("2. gallons to liters")

            sub = input("Choose option: ")

            if sub == "1":
                result = value * 0.264172
                print("Result:", result, "gal")

            elif sub == "2":
                result = value / 0.264172
                print("Result:", result, "L")

        else:
            print("Invalid conversion type")

    except ValueError:
        print("Invalid input. Please enter numeric values.")

unit_converter()
```

**Task 2 — Mathematical Operations on List**

```python
def calculate(numbers, operation):
    if operation == "sum":
        return sum(numbers)
    elif operation == "avg":
        return sum(numbers) / len(numbers)
    elif operation == "max":
        return max(numbers)
    elif operation == "min":
        return min(numbers)

try:
    numbers = [float(x) for x in input("Enter numbers separated by spaces: ").split()]

    operations = {
        "1": "sum",
        "2": "avg",
        "3": "max",
        "4": "min"
    }

    print("\nSelect operation")
    print("1 - Sum")
    print("2 - Average")
    print("3 - Maximum")
    print("4 - Minimum")

    choice = input("Enter choice: ")

    if choice in operations:
        result = calculate(numbers, operations[choice])
        print("Result:", result)
    else:
        print("Invalid choice")

except ValueError:
    print("Please enter valid numbers.")
except ZeroDivisionError:
    print("Cannot calculate average of empty list.")
```

# **Exercise on List Manipulation:**

**1 Extract every other element**

```python
def extract_every_other(lst):
    result = []
    for i in range(0, len(lst), 2):
        result.append(lst[i])
    return result

print(extract_every_other([1,2,3,4,5,6]))
```

**2 Get sublist**

```python
def get_sublist(lst, start, end):
    """Returns sublist from start to end index inclusive."""
    return lst[start:end+1]

print(get_sublist([1,2,3,4,5,6],2,4))
```

**3 Reverse list**

```python
def reverse_list(lst):
    return list(reversed(lst))

print(reverse_list([1,2,3,4,5]))
```

**4 Remove first and last**

```python
def remove_first_last(lst):
    """Removes first and last element."""
    lst.pop(0)
    lst.pop(-1)
    return lst

print(remove_first_last([1,2,3,4,5]))
```

**5 Get first n**

```python
def get_first_n(lst, n):
    result = []
    for i in range(n):
        result.append(lst[i])
    return result

print(get_first_n([1,2,3,4,5],3))
```

**6 Get last n**

```python
def get_last_n(lst, n):
    """Returns last n elements."""
    return lst[-n:]

print(get_last_n([1,2,3,4,5],2))
```

**7 Reverse skip**

```python
def reverse_skip(lst):
    result = []
    for i in range(len(lst)-2, -1, -2):
        result.append(lst[i])
    return result

print(reverse_skip([1,2,3,4,5,6]))
```

# **Exercise on Nested List:**

**1 Flatten list**

```python
def flatten(lst):
    return [item for sub in lst for item in sub]

print(flatten([[1,2],[3,4],[5]]))
```

**2 Access nested element**

```python
from functools import reduce

def access_nested_element(lst, indices):
    return reduce(lambda x, y: x[y], indices, lst)

lst = [[1,2,3],[4,5,6],[7,8,9]]
print(access_nested_element(lst, [1,2]))
```

**3 Sum nested**

```python
def sum_nested(lst):
    """Returns sum of all elements in nested list."""
    total = 0
    for item in lst:
        if isinstance(item,list):
            total += sum_nested(item)
        else:
            total += item
    return total

print(sum_nested([[1,2],[3,[4,5]],6]))
```

**4 Remove element**

```python
def remove_element(lst, elem):
    return [remove_element(i, elem) if isinstance(i, list) else i
            for i in lst if i != elem]

print(remove_element([[1,2],[3,2],[4,5]],2))
```

**5 Find max**

```python
def find_max(lst):
    """Finds maximum in nested list."""
    maximum = float('-inf')
    for item in lst:
        if isinstance(item,list):
            maximum = max(maximum, find_max(item))
        else:
            maximum = max(maximum,item)
    return maximum

print(find_max([[1,2],[3,[4,5]],6]))
```

**6 Count occurrences**

```python
def count_occurrences(lst, elem):
    return sum(count_occurrences(i, elem) if isinstance(i, list) else (i == elem) for i in lst)

print(count_occurrences([[1,2],[2,3],[2,4]], 2))
```

**7 Deep flatten**

```python
def deep_flatten(lst):
    if not lst:
        return []
    if isinstance(lst[0], list):
        return deep_flatten(lst[0]) + deep_flatten(lst[1:])
    return [lst[0]] + deep_flatten(lst[1:])

print(deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
```

**8 Nested average**

```python
def average_nested(lst):
    """Returns average of nested list."""
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)

print(average_nested([[1,2],[3,4],[5,6]]))
```

```python
import numpy as np

# 1. Array from 10 to 49
arr = np.arange(10, 50)
print("10 to 49:\n", arr)

# Alternative: using np.linspace
arr_alt = np.linspace(10, 49, 40, dtype=int)
print("\n10 to 49 (alternative):\n", arr_alt)

# 2. 3x3 matrix from 0 to 8
matrix = np.arange(9).reshape(3,3)
print("\n3x3 Matrix:\n", matrix)

# Alternative: using np.arange + reshape with a different syntax
matrix_alt = np.arange(0, 9).reshape(3, 3)
print("\n3x3 Matrix (alternative):\n", matrix_alt)

# 3. Identity matrix
identity = np.eye(3)
print("\nIdentity Matrix:\n", identity)

# Alternative: using np.identity
identity_alt = np.identity(3)
print("\nIdentity Matrix (alternative):\n", identity_alt)

# 4. Random array of size 30 and mean
rand_arr = np.random.rand(30)
print("\nRandom Array Mean:", rand_arr.mean())

# Alternative: using np.random.uniform
rand_arr_alt = np.random.uniform(0, 1, 30)
print("Random Array Mean (alternative):", rand_arr_alt.mean())

# 5. 10x10 random array with min and max
rand_matrix = np.random.rand(10, 10)
print("\nMinimum:", rand_matrix.min())
print("Maximum:", rand_matrix.max())

# Alternative: using np.random.uniform
rand_matrix_alt = np.random.uniform(size=(10,10))
print("Minimum (alternative):", rand_matrix_alt.min())
print("Maximum (alternative):", rand_matrix_alt.max())

# 6. Zero array size 10 and replace 5th element
zero_arr = np.zeros(10)
zero_arr[4] = 1
print("\nZero array with 5th element replaced:\n", zero_arr)

# Alternative: using np.put
zero_arr_alt = np.zeros(10)
np.put(zero_arr_alt, 4, 1)
print("\nZero array with 5th element replaced (alternative):\n", zero_arr_alt)

# 7. Reverse array
arr = np.array([1,2,0,0,4,0])
print("\nReversed array:\n", arr[::-1])

# Alternative: using np.flip
arr_alt = np.flip(arr)
print("Reversed array (alternative):\n", arr_alt)

# 8. 2D array with border 1 and inside 0
border = np.ones((5,5))
border[1:-1,1:-1] = 0
print("\nBorder array:\n", border)

# Alternative: using np.pad
inner = np.zeros((3,3))
border_alt = np.pad(inner, pad_width=1, constant_values=1)
print("Border array (alternative):\n", border_alt)

# 9. Checkerboard pattern
checker = np.zeros((8,8))
checker[1::2,::2] = 1
checker[::2,1::2] = 1
print("\nCheckerboard pattern:\n", checker)

# Alternative: using np.indices and sum modulo 2
checker_alt = (np.indices((8,8)).sum(axis=0) % 2)
print("Checkerboard pattern (alternative):\n", checker_alt)
```

**Problem 4 — Matrix Operations**

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A * A^-1 = I
A_inv = np.linalg.inv(A)
result = A @ A_inv   # using @ instead of np.dot
print("A * A^-1 =\n", result)

# 2. Prove AB ≠ BA
AB = A @ B
BA = B @ A
print("\nAB =\n", AB)
print("\nBA =\n", BA)
print("\nAB == BA? ", np.allclose(AB, BA))  # checks equality numerically

# 3. Prove (AB)^T = B^T A^T
left = (AB).T
right = B.T @ A.T
print("\n(AB)^T =\n", left)
print("\nB^T A^T =\n", right)
print("\nAre they equal? ", np.allclose(left, right))
```

**Matrix Multiplication**

```python
n = 50

A = np.random.rand(n,n).tolist()
B = np.random.rand(n,n).tolist()

start = time.time()

result = [[0]*n for _ in range(n)]

for i in range(n):
    for j in range(n):
        for k in range(n):
            result[i][j] += A[i][k] * B[k][j]

end = time.time()

print("Python Matrix Multiplication Time:", end - start)
```

```python
A = np.random.rand(n,n)
B = np.random.rand(n,n)

start = time.time()

result = np.dot(A, B)

end = time.time()

print("NumPy Matrix Multiplication Time:", end - start)
```

NumPy operations are significantly faster than Python loops because NumPy uses vectorization and optimized C implementations. Python lists rely on explicit loops which are slower. Therefore, for large numerical computations, NumPy provides better performance and efficiency.
